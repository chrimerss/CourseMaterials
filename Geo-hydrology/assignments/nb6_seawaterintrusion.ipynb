### Seawater intrusion
Groundwater flow in coastal aquifers contains both freshwater and saltwater. 
The saltwater originates from seawater and contains different salts, of which NaCl is the most abundant one. Chloride concentrations in seawater are $\sim$15-20 g/L, while water with a chloride concentration of 0.25 g/L already tastes salty. 
Seawater is heavier than the freshwater due to the presence of the salt. 
 The density of freshwater is $\rho\approx 1000$ kg/m$^3$ while the density of seawater is $\rho\approx 1025$ kg/m$^3$.  Note that the saltwater in coastal aquifers isn't all that salty: the density of a saturated salt solution is on the order of 1200 kg/m$^3$.
 
 The density of groundwater is a function of many parameters, including the pressure, temperature and salinity. In coastal aquifers, the effect of the salinity is the dominant one. The relationship between the salinity and the density is called an equation of state and can be approximated for salt groundwater originating from seawater by (e.g., Oude Essink et al. 2010)

$$
\rho = \rho_f(1+\beta C)
$$

where  $C$ is the chloride concentration in mg/L and $\beta \approx 1.34 Ã— 10^{-6}$ L/(mg Cl) is the volumetric concentration expansion gradient  (Oude Essink et al., 2010).

Freshwater floats on top of saltwater as the freshwater is lighter than the saltwater. The freshwater is seperated from the saltwater by a brackish transition zone in which the water varies from freshwater to saltwater.
The transition zone between freshwater and saltwater is often thin, in which case it can be represented by an interface separating the freshwater from the saltwater. The landward intrusion of seawater in coastal aquifers is referred to as seawater intrusion where saltwater forms a wedge below the freshwater (e.g., Werner et al., 2013). The point of the wedge along the base of the aquifer that intrudes farthest inland is referred to as the toe of the interface. The toe moves farther inland when the flux of freshwater towards the coast decreases.

### Heads in coastal aquifers
The definition of head is the elevation to which water rises in a standpipe (an observation well) and may be computed from

$$
h = \frac{p}{\rho g} + z
$$

where $p$ is the pressure in the groundwater, $z$ is the elevation where the pressure is meaured, and $\rho$ is the density.
In coastal aquifers, this definition does not suffice as the density varies (e.g., Post et al., 2007). When the freshwater density is used, the head is referred to as the freshwater head

$$
h =  \frac{p}{\rho_f g} + z \qquad \text{freshwater head}
$$

where $\rho_f$ is the density of freshwater.
The freshwater head is the elevation to which water rises in a standpipe when the standpipe is filled with freshwater. 
In a similar fashion, the saltwater head $h_s$ is defined as the elevation to which water rises in a standpipe when the standpipe is filled with saltwater. 

$$
h_s =  \frac{p}{\rho_s g} + z \qquad \text{saltwater head}
$$

where $\rho_s$ is the density of saltwater.
Finally, the pointwater head $h_p$ is defined as the elevation to which water rises in a standpipe when the standpipe is filled with water that exists in the aquifer at the bottom of the standpipe. 

$$
h_p =  \frac{p}{\rho_p g} + z \qquad \text{pointwater head}
$$

where $\rho_p$ is the density of the water that exists at the bottom of the standpipe. The pointwater head is not very convenient for doing calculations, but it is the head that is commonly measured in the field. The procedure for measuring the head in an observation well in a coastal aquifer is to pump a little water from the well (at least the volume of the observation well) so that the density in the observation well is the same everywhere, after which both the (pointwater) head (or pressure) and density need to be recorded. 

Calculations are commonly done with freshwater heads, so measured pointwater heads need to be converted to freshwater heads. Consider a pointwater head $h_p$ measured at elevation $z_p$ and the density in the observation well is $\rho_p$. The pressure $p$ at the bottom of the observation well, with elevation $z_w$ is

$$
p = \rho_pg(h_p-z_w)
$$

The equivalent freshwater head needs to result in the same pressure $p$ when the observation well is filled with freshwater

$$
\rho_pg(h_p-z_w) = \rho_fg(h-z_w)
$$

which means that

$$
h = z_w + \frac{\rho_p}{\rho_f}(h_p-z_w)
$$

where it is important to realize that both the heads and the elevation of the bottom of the observation well $z_w$ are measured with respect to the same datum.

### Darcy's law
The flow of groundwater through a pore space is accompanied by a friction force between the flowing fluid and the soil skeleton (Verruijt, 2012). The equations of equilibrium are 

$$\begin{split}
&\frac{\partial p}{\partial x}-f_x=0 \\
&\frac{\partial p}{\partial y}-f_y=0 \\
&\frac{\partial p}{\partial z}+\rho g - f_z =0
\end{split}$$

where $f_x$, $f_y$ and $f_z$ are the components of the force, per unit volume, onto the soil skeleton and vice versa, and $\rho$ is the density of the water. It is reasonable to assume a linear relation between the force and the specific discharge factor

$$\begin{split}
&f_x = -\frac{\mu}{\kappa} q_x \\
&f_y = -\frac{\mu}{\kappa} q_y \\
&f_z = -\frac{\mu}{\kappa} q_z 
\end{split}$$

where, as defined in Notebook 1, $\kappa$ [L$^2$] is the intrinsic permeability of the
aquifer and $\mu$ [MLT$^{-1}$] is the dynamic viscosity.
Substitution of these forces into the previous equations directly leads to Darcy's law in terms of pressure.

$$\begin{split}
q_x&= -\frac{\kappa}{\mu}\frac{\partial p}{\partial x} \\
q_y&= -\frac{\kappa}{\mu}\frac{\partial p}{\partial y} \\
q_z&= -\frac{\kappa}{\mu}\frac{\partial p}{\partial z} -\frac{\kappa}{\mu}\rho g
\end{split}$$

where the density may vary with time and space $\rho=\rho(x,y,z,t)$. Using the freshwater head $h$

$$
h = \frac{p}{\rho_f g} + z
$$

or 

$$
p=\rho_f g(h-z)
$$

Darcy's law becomes

$$\begin{split}
q_x&= -k\frac{\partial h}{\partial x} \\
q_y&= -k\frac{\partial h}{\partial y} \\
q_z&= -k\frac{\partial h}{\partial z} +k\frac{\rho_f}{\rho_f}-k\frac{\rho}{\rho_f}=-k\left[\frac{\partial h}{\partial z} +\nu\right]
\end{split}$$

where $k=\kappa\rho_f g/\mu$ is the hydraulic conductivity of fresh water and 

$$\nu=(\rho-\rho_f)/\rho_f$$ 

is the dimensionless density difference. Note that in freshwater, $\nu=0$ and $q_z$ reduces to the common $q_z=-k\partial h/\partial z$. 

Contours plots of the freshwater head in a horizontal plane are useful, as the flow is always normal to contours of the freshwater head in a horizontal plane, even when the density varies. Contours of the freshwater head in a vertical plane are not very useful when the density varies, as flow is not normal to contours of the freshwater head in a vertical plane when the density varies. 
The choice of formulating Darcy's law in terms of the freshwater head is an arbitrary one, although convenient as large portions of a coastal aquifer commonly consist of freshwater, where the use of the freshwater head is an obvious one. 

The consequence of using freshwater heads in stagnant saltwater is that the freshwater head increases with depth while the flow is zero. This can be seen directly from the equation for $q_z$, where $q_z=0$ means that $\partial h/\partial z = -\nu$. So when $\nu>0$, then the vertical gradient of the freshwater head is negative and the freshwater head increases with depth.

Alternatively, any other constant density can be used to convert Darcy's equation from pressure to head. For example, when using the density of saltwater, the vertical flow becomes

$$\begin{split}
q_x&= -k_s\frac{\partial h_s}{\partial x} \\
q_y&= -k_s\frac{\partial h_s}{\partial y} \\
q_z&= -k_s\left[\frac{\partial h_s}{\partial z} +\nu_s\right]
\end{split}$$

where $h_s$ is the saltwater head, $k_s=\kappa\rho_sg/\mu$, and $\nu_s=(\rho-\rho_s)/\rho_s$. Note that in the saltwater zone $\nu_s=0$, and that stagnant saltwater is equivalent to a constant saltwater head $h_s$ everywhere.

*Darcy's law in terms of the freshwater head will be used in the remainder of this Notebook.*

### Depth of the interface
Consider steady interface flow in a coastal aquifer. The saltwater is at rest, which means that the saltwater head $h_s$ is constant everywhere in the saltwater zone. Consider two observation wells right next to eachother. The first observation well has its opening just above the interface and is filled with freshwater. The head is equal to the head in the freshwater zone at that point. The observation well next to it has its opening just below the interface and is filled with saltwater. The saltwater head in the latter well is equal to sealevel, since the saltwater is at rest. The interface is a distance $d$ below sealevel. At the interface, the pressure $p_f$ in the freshwater must be equal to the pressure $p_s$ in the saltwater

$$
p_f = p_s 
$$

When the head $h$ is measured with respect to sealevel, the pressure at the bottom of the freshwater well is the height of the column of freshwater multiplied with $\rho_fg$ and the pressure at the bottom of the saltwater well is the height of the column of saltwater multiplied with $\rho_sg$ 

$$
(h+d)\rho_fg = d\rho_sg
$$

The distance $d$ of the interface below sealevel may now be computed as

$$
d = \frac{\rho_f}{\rho_s-\rho_f}h=\alpha h
$$

where $\alpha=1/\nu$
This equation is known as the equation of Bhadon-Ghyben-Herzberg. For seawater with a density of $\rho_s=1025$ kg/m$^3$, $\nu=0.025$, which means that $d=40h$. In words: when the head in the freshwater is $h$ above sealevel, the interface is at a depth of $40h$ below sealevel *when flow is steady and the seawater is at rest*.

### Discharge potential for steady interface flow
Consider steady interface flow in a confined aquifer. The freswater is flowing and the saltwater is at rest. The Dupuit approximation is adopted for flow in the freshwater zone, which means that the resistance to vertical flow in the freshwater zone is neglected and the freshwater head is a function of the horizontal $x$ and $y$ coordinates only. Flow may be formulated using discharge potentials (Strack, 1976).

#### Steady unconfined interface flow
Consider steady unconfined interface flow in a coastal aquifer. The freshwater head is measured with respect to sealevel. The depth of the interface is equal to $d=\alpha h$, so that the total thickness of the freshwater zone is equal to $h+d=(1+\alpha)h$. The discharge vector in the freshwater zone may be written as

$$
\vec{Q} = -k(\alpha+1)h\vec{\nabla} h= -\vec{\nabla}\Phi
$$

The discharge potential for unconfined interface flow is defined as

$$
\Phi = \tfrac{1}{2}k(\alpha+1)h^2
$$

The head at the toe of the interface is $h_\text{toe}=D/\alpha$, where $D$ is the depth of the bottom of the aquifer below sealevel. The potential at the toe of the interface is

$$
\Phi_\text{toe} = \tfrac{1}{2}k\frac{\alpha+1}{\alpha^2}D^2
$$

<img src="unconfined_interface.png" width="400">
**Figure** Unconfined interface flow.

Upstream of the toe, the discharge vector for regular unconfined flow may be written as

$$
\vec{Q} = -k(h+D)\vec{\nabla} h = -\vec{\nabla}\Phi
$$

where the discharge potential is defined as

$$
\Phi = \tfrac{1}{2}k(h+D)^2 + C
$$

The constant $C$ is chosen such that the discharge potential at the toe is equal to the discharge potential for interface flow

$$
\frac{1}{2}k\left(\frac{D}{\alpha}+D\right)^2 + C=\tfrac{1}{2}k\frac{\alpha+1}{\alpha^2}D^2
$$

so that 

$$
C = -\tfrac{1}{2}k\frac{\alpha+1}{\alpha}D^2
$$

### Example

Consider uniform flow towards the coast at a rate $Q_0$ as shown in the figure above, so that $Q_x=-Q_0$. Given: $k=10$ m/d, $D=20$ m, $\rho_s=1025$ kg/m$^3$, $Q_0=0.4$ m$^2$/d. Question: where it the toe of the interface?

The potential in the aquifer is (since the potential along the coast ($x=0$) is zero)

$$
\Phi = Q_0x
$$

The location of the toe is found at the position where 
$\Phi=\Phi_\text{toe}$, which gives 

$$
x_\text{toe}=\tfrac{1}{2}k\frac{\alpha+1}{\alpha^2Q_0}D^2=128 \quad \text{m}
$$

The head and position of the interface may be plotted for $x$ between 0 and 200 as follows

```python
k = 10
D = 20
rhof = 1000
rhos = 1025
Q0 = 0.4
alpha = rhof / (rhos - rhof)

def head(x, Q0):
    phitoe = 0.5 * k * (1 + alpha) * D ** 2 / alpha ** 2
    C = -0.5 * k * D ** 2 * (1 + alpha) / alpha
    phi = Q0 * x
    h = zeros_like(phi)
    h[phi < phitoe] = sqrt(2 * phi[phi < phitoe] / (k * (1 + alpha)))
    h[phi > phitoe] = sqrt(2 * (phi[phi >= phitoe] - C) / k) - D
    return h

x = linspace(0, 200, 100)
h = head(x, Q0)
zeta = -alpha * h
zeta[zeta < -D] = -D
plot(x, h, 'b')
plot(x, zeta, 'r')
plot(x, -D * ones_like(x), 'k');
```

### Steady confined flow
Consider steady confined flow in a coastal aquifer. The freshwater head is measured with respect to sealevel. The top of the confined aquifer is a distance $D$ below sealevel. The thickness of the confined aquifer is $H$. The depth of the interface below sealevel is $d=\alpha h$, so that the total thickness of the freshwater zone is equal to $d-D=\alpha h - D$. The discharge vector in the freshwater zone may be written as 

$$
\vec{Q} = -k(\alpha h-D)\vec{\nabla} h = = -k\alpha(h-D/\alpha)\vec{\nabla} h = -\vec{\nabla}\Phi
$$

The discharge potential for confined interface flow is defined as

$$
\Phi = \tfrac{1}{2}k\alpha(h-D/\alpha)^2
$$

The head at the toe of the interface is $h_\text{toe}=(D+H)/\alpha$ so that the potential at the toe of the interface is

$$
\Phi_\text{toe} = \tfrac{1}{2}k\alpha[(D+H)/\alpha - D/\alpha]^2 = \tfrac{1}{2}kH^2/\alpha
$$

<img src="confined_interface.png" width="400">
**Figure** Confined interface flow.

Upstream of the toe, the discharge vector for regular confined flow may be written as 

$$
\vec{Q} = -kH\vec{\nabla} h = -\vec{\nabla}\Phi
$$

where the discharge potential is defined as

$$
\Phi = kHh + C
$$

The constant $C$ is chosen such that the discharge potential at the toe is equal to the discharge potential for interface flow

$$
\tfrac{1}{2}k\frac{H^2}{\alpha} = kH\frac{D+H}{\alpha} + C
$$

so that 

$$
C = -\frac{k(DH+\tfrac{1}{2}H^2)}{\alpha}
$$

### Vertical flow when the density varies vertically
Consider steady flow in a homogeneous aquifer with vertically varying density. The horizontal component of the specific discharge vector is constant, so that the differential equation for the vertical component of flow may be written is

$$
\frac{\text{d} q_z}{\text{d} z} = 0
$$

Application of Darcy's law in terms of freshwater head gives

$$
\frac{\text{d}^2 h}{\text{d} z^2} + \frac{\text{d}\nu}{\text{d}z} = 0
$$

First, consider vertical flow with the following boundary conditions

$$\begin{split}
&z=H \qquad h=h_2 \qquad \nu=\nu_2 \\
&z=0 \qquad h=h_1 \qquad \nu=\nu_1
\end{split}$$

and a dimensionless density $\nu$ that varies linearly from $z=0$ till $z=H$

$$
\nu = \frac{\nu_2 - \nu_1}{H}z + \nu_1 \qquad \frac{\text{d}\nu}{\text{d}z} = \frac{\nu_2-\nu_1}{H}
$$

The solution to the differential equation for $h$ given $\nu$ and the boundary conditions defined above gives

$$
h = -\frac{\nu_2-\nu_1}{2H}z^2 + \left( \frac{h_2-h_1}{H} + \frac{\nu_2-\nu_1}{2} \right) z + h_1
$$

so that the vertical flux is

$$
q_z = -k\left[ -\frac{\nu_2-\nu_1}{H}z + \frac{h_2-h_1}{H} + \frac{\nu_2-\nu_1}{2} + \frac{\nu_2-\nu_1}{H}z + \nu_1 \right]
$$

which simplifies to

$$
q_z = -k\left[ \frac{h_2-h_1}{H} + \frac{1}{2}(\nu_1+\nu_2) \right]
$$

Hence, $q_z$ may be computed as $k$ times the difference in freshwater head divided by the distance between them plus the average dimensionless density between the top and bottom.

Next, we solve the same differential equation for the head but this time the variation of the density is an arbitrary function $\nu(z)$:

$$\begin{split}
&z=H \qquad h=h_2 \qquad \nu=\nu(H) \\
&z=0 \qquad h=h_1 \qquad \nu=\nu(0)
\end{split}$$

Integration of the differential equation twice gives

$$
h = -\int_0^z \nu\text{d}z + Az + B
$$

If we write the integral as the function $N(z)$

$$
N(z) = \int_0^z \nu\text{d}z
$$

and apply the boundary conditions, the solution becomes

$$
h = -N(z) + \frac{h_2-h_1+N(H)}{H}z + h_1
$$

The corresponding vertical flow is

$$
q_z = -k\left[-\nu + \frac{h_2-h_1+N(H)}{H} + \nu\right]
$$

which simplifies to

$$
q_z = -k\left[\frac{h_2-h_1}{H}+\frac{1}{H} \int_0^H \nu\text{d}z\right] = -k\left[\frac{h_2-h_1}{H}+\overline{\nu} \right]
$$

where $\overline{\nu}$ is the average dimensionless density between $z=0$ and $z=H$. In conclusion, the vertical flux between two points only depends on the difference in freshwater head divided by the distance between them plus the average dimensionless density difference between the two points. 

This holds even if the density distribution is discontinuous. Consider a layer with thickness $H$ consisting of two parts, the first part with thickness $H_a$ and average density difference $\overline{\nu}_a$, and the second part with thickness $H_b$ and average density difference $\overline{\nu}_b$. If the head at the boundary between sections $a$ and $b$ is $h^*$, then the flux in both zones may be written as

$$\begin{split}
&q_z = -k\left[\frac{h^*-h_1}{H_a}+\overline{\nu}_a \right] \qquad \text{zone $a$} \\
&q_z = -k\left[\frac{h_2-h^*}{H_b}+\overline{\nu}_b \right] \qquad \text{zone $b$}
\end{split}$$

Elimination of $h^*$ from both equations gives

$$
q_z = -k\left[\frac{h_2-h_1}{H}+\frac{H_a\overline{\nu}_a + H_b\overline{\nu}_b}{H} \right]
$$

so that again, the last term represents the average density difference in the layer.
