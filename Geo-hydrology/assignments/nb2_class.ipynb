```python
x, y = np.meshgrid(np.linspace(-200, 200, 100), np.linspace(-100, 100, 100))
```

```python
r1 = sqrt((x + d) ** 2 + y ** 2)
r2 = sqrt((x - d) ** 2 + y ** 2)
phi = Q / (2 * pi) * log(r1 / r2) + phi0
h = phi / (k * H)
```

```python
plt.contour(x, y, h, np.linspace(37, 43, 61))
plt.axis('equal')
```

```python
plot(x[0], h[50])
```

```python
Psi = Q / (2 * pi) * (np.arctan2(y, x + d) - np.arctan2(y, x - d))
plt.contour(x, y, phi, 20, colors='b')
plt.contour(x, y, Psi, 20, colors='r', linestyles='-')
plt.axis('scaled')
plt.xlim(-200, 0)
```
